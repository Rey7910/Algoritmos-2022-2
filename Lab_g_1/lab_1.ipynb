```python
def busqueda_subcadena(texto, subcadena):
  ''' Esta funci√≥n retorna una tupla de la forma (int repeticiones - lista de index)
      Repeticiones hace referencia a la cantidad de repeticiones que aparece la
      subcadena en texto.
      Lista de index hace referencia a una lista que contiene tuplas de la forma
      (INDICE INICIAL  -  INDICE FINAL) 
    '''
  index = 0
  repeticiones = 0
  
  apariciones = []

  for i in texto:
    if i == subcadena[0] and texto[index:len(subcadena)+index] == subcadena: 
      apariciones.append((index, index+len(subcadena)-1 ))
      repeticiones = repeticiones + 1
    
    index = index+1
  return (repeticiones, apariciones)


texto = ""

#cadena objetivo
cadena = "ACATTACA"

#en resultado se almacena la tupla (cantidad, indices)
resultado = busqueda_subcadena(texto,cadena)

#aqu√≠ se imprimen los valores de resultado
print(resultado[0])
print(resultado[1])


#Aqu√≠ se imprimen los indices, se suma uno en el print del slicing para ampliar el rango y que quede toda la cadena
for i in resultado[1]:
  print(texto[i[0]:i[1]+1])

```
