```python

# Python program to demonstrate searching operation
# in binary search tree without recursion

element = []
levels=[]

class newNode:
 
    # Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
 
# Function to check the given
# key exist or not
def iterativeSearch(root, key):
     
    # Traverse until root reaches
    # to dead end
    
    location_array=[]
    while root != None:
         
        # pass right subtree as new tree
        if key > root.data:
            
            location_array.append("Derecha")
           
            root = root.right
 
        # pass left subtree as new tree
        elif key < root.data:
            location_array.append("Izquierda")
            root = root.left
        else:
            print(location_array)
            return True # if the key is found return 1
    return False
 
# A utility function to insert a
# new Node with given key in BST
def insert(Node, data):
     
    # If the tree is empty, return
    # a new Node
    if Node == None:
        return newNode(data)
 
    # Otherwise, recur down the tree
    if data < Node.data:
        
        Node.left = insert(Node.left, data)
    elif data > Node.data:
        Node.right = insert(Node.right, data)
 
    # return the (unchanged) Node pointer
    return Node
    
def preorder_keeper(root,level):


    
    levels.append(level)
    element.append(root.data)

    level_right=level
    level_left = level
    
    if(root.left!=None):
        level_left+=1
        preorder_keeper(root.left,level_left)
    
    if(root.right!=None):
        level_right+=1
        preorder_keeper(root.right,level_right)
        
   
    
def draw_tree():
        
        maxi=max(levels)
        
        
        top_maxi=maxi
      
        iterations = (9*maxi)
        
        if(iterations%2==0):
            iterations+=1
        
        separator = (3*5)*maxi
        
        level = 0
        tree_level=0
        
        delimiter=4+maxi
        depth=0
        
        
        for i in range(iterations):
            
            if(maxi>5):
                print("El árbol tiene demasiados niveles para ser dibujado")
                break;
            
            
            if(i%delimiter==0):
                #print("n")
                
                tree_level+=1
                
                
                    
                
                if(i==0):
                    level+=1
                else:
                    level*=2
                   
                empty_spaces  = " "*separator
                print(empty_spaces, end = "") 
                
                second=0
                node1 = "Not"
                node2 = "Not"
                controler=0
                controler_2=0
                
                for j in range(level):
                    controler=0
                    contoler_2=0
                    
                    for y in range(len(levels)):
                        
                        if(levels[y]==tree_level and tree_level==1):
                            print(element[y],end="")
                            levels[y]=100
                            break;
                        
                        else:
                            
                            
                            
                            if(levels[y]==tree_level):
                                
                                if(node1!="Not" and node2=="Not"):
                                    node2 = element[y]
                                    
                                    if(element[y]>100): 
                                        controler_2+=2
                                    elif(element[y]>10):
                                        controler_2+=1
                                    
                                    levels[y]=100
                                    second+=1
                                    
                                elif(node1=="Not"):
                                    node1= element[y]
                                    
                                    if(element[y]>100): 
                                        controler+=2
                                    elif(element[y]>10):
                                        controler+=1
                                        
                                    levels[y]=100
                                    second+=1
                                    
                            
                            if(second==2 or y==len(levels)-1 and node1!="Not"):
                                
                                second=0
                                numbers=str(node1) +" "*(((maxi*depth)*2)-controler)+str(node2)
                                
                                node1 = "Not"
                                node2 = "Not"
                                controler+=controler_2
                                
                                print(numbers,end="")
                                
                                if(top_maxi==3 or top_maxi==2):
                                    empty_spaces = " "*((((delimiter+1)*3)*maxi)-(maxi*depth)*2)
                                elif(top_maxi==4):
                                    empty_spaces = " "*(((delimiter+10)*maxi)-(maxi*depth)*2)
                            
                                elif(top_maxi==5):
                                    
                                    if(delimiter==6):
                                       empty_spaces = " "*((13*maxi)-(maxi*depth)*2)
                                    else:
                                        
                                       empty_spaces = " "*((((delimiter*3)-3)*maxi)-(maxi*depth)*2)
                                   
                                print(empty_spaces,end="")
                    
                            
                            
                    
                    
                
                print("\n",end="")
                
                maxi-=1
                depth=0
                delimiter-=1
                
                if(maxi==0):
                    break;
            
            else:
                empty_spaces  = " "*separator
                print(empty_spaces, end = "")
                
                for j in range(level):
                    sticks="/"+" "*(maxi*depth)*2+"\\"
                    last_depth=depth
                    
                    
                    if(j==level-1):
                        print(sticks)
                        
                    else:
                        print(sticks,end="")
                        
                        
                        if(top_maxi==3 or top_maxi==2):
                            empty_spaces = " "*((((delimiter+1)*3)*maxi)-(maxi*depth)*2)
                        elif(top_maxi==4):
                            empty_spaces = " "*(((delimiter+10)*maxi)-(maxi*depth)*2)
                        
                        elif(top_maxi==5):
                            
                            if(delimiter==6):
                               empty_spaces = " "*((13*maxi)-(maxi*depth)*2)
                            else:
                                
                               empty_spaces = " "*((((delimiter*3)-3)*maxi)-(maxi*depth)*2)
                            
                        
                        print(empty_spaces,end="")
                  
                    
                    
                depth+=1        
               

                separator-=maxi
                last_depth=depth
                
```
